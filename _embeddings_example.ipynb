```python
from transformers import BertTokenizer
tokenizer = BertTokenizer.from_pretrained('google-bert/bert-base-uncased')

from llama_index.embeddings.text_embeddings_inference import (
    TextEmbeddingsInference,
)
from llama_index.core.embeddings import resolve_embed_model
embed_model = resolve_embed_model("local:BAAI/bge-small-en-v1.5")

def get_embeddings(model, query):
    embeddings = model.get_text_embedding(query)
    print(f"Embedding: {query}")
    print(embeddings)

def tokenize(sentence):

    print(f"{sentence}: ", tokenizer.encode(sentence))
```

```python

tokenize('Josh is very tall')

tokenize("Corinne is very tall")

tokenize("Muppets live on sesame street")
```

```python


get_embeddings(embed_model, "Josh is very tall")

get_embeddings(embed_model, "Corinne is very tall")
```
